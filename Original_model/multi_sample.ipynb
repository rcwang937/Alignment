```python
model = AutoModelForCausalLM.from_pretrained("GAIR/confucius-multisample", use_auth_token=True)
```

```python
trivia_eval = load_dataset("trivia_qa", "unfiltered", split="validation")
```

```python
# format the prompt for the model
def format_confidence_verb_prompt(question):
    return f"Answer the question. If you don’t know the answer to the question, it is appropriate to say “I apologize, but I’m not able to provide an answer to the question.\nQ: {question}\nA:"


# Function to generate a response and extract confidence
def get_model_response(question):
    inputs = tokenizer(format_confidence_verb_prompt(question), return_tensors="pt")
    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=150)

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
```

```python
# Evaluate the model on a subset of TriviaQA for example purposes
results = []
num_samples = 50  # You can increase this to evaluate on more samples

for i in range(num_samples):
    question = trivia_eval[i]["question"]
    gold_answer = trivia_eval[i]["answer"]["value"]

    # Get the model's response
    response = get_model_response(question)

    # Store the result with question, gold answer, and model response
    results.append({
        "question": question,
        "gold_answer": gold_answer,
        "model_response": response
    })
    print(results)
```
